```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
table = pd.read_csv('songs_normalize.csv', encoding='latin_1')
table
```

```python
table = table.drop(table[table['year'] == 1998].index)
table = table.drop(table[table['year'] == 2020].index)
```

```python
year_with = table.drop_duplicates(subset=['year'])
year_com = year_with.sort_values(by='year')
mean_pop = table['popularity'].mean()
a1 = year_com.count()
spisok = [i for i in range(1, a1['year'] + 1)]
spisok2 = [i for i in range(0, 110, 10)]
spisok
```

```python
pop_main = []
pop_main1 = []
years = a = table['year'].min()
for i in range(21):
  mask = table[table.year == years]
  mean_pop1 = mask['popularity'].mean()
  pop_main += [mean_pop1]
  pop_main1 += [years]
  years += 1
```

```python
figura = plt.figure(figsize = (18, 15))
ax = figura.add_subplot(1,1,1)
ax.set_title('Рейтинг песен с 1998 по 2020 года')
ax.plot(spisok, pop_main, color = 'purple', linestyle='-.', marker='*')
ax.set_xticks(spisok)
ax.set_yticks(spisok2)
x = ax.set_xticklabels(year_com['year'], rotation =90)
ax.grid()
```
